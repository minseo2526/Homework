```python
frequency={}
while True:
  c=input('알파벳 1자응 입력하세요[끝내려면 엔터를 치세요]:')
  if c=='':
    break
  if c not in frequency:
    frequency[c]=1
  else:
    frequency[c]+=1
  print(frequency)
```

```python
greats={'유비':89,'관우':95,'장비':91,'조운':94}
for name in iter(greats):
  print(name)
```
