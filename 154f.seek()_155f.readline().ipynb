```python
with open('prophet_en.txt','r') as f:
  line=f.readline()
  while line != '':
    print(f'>>{line[:-1]}')
    k=input('다음 줄:엔터 입력.종료/q 입력:')
    if k=='/q':
      break
    line=f.readline()
```

```python
with open('prophet_en.txt','r') as f:
  line=f.readlines()
  for l in line: # Changed 'lines' to 'line' and 'line' to 'l' to avoid shadowing
    print(f'>>{l[:-1]}')
    k=input('다음 줄:엔터 입력.종료/q 입력:')
    if k=='/q':
      break
```
