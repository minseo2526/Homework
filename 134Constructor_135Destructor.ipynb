```python
class Wizard:
  def __init__(self,name=None):
    self.name=name
    print(f'마법사 이름이 <{name}>인 캐릭터가 생성되었습니다.')

  def __del__(self):
    print('Wizard 객체가 메모리에서 제거됩니다.')
```

```python
player1=Wizard('StormBaby')
del player1
```
