```python
from os import rename

src='sunrise.jpg' # Changed from 'sunrise.copied_jpg'
dst='sunrise2.jpg'
rename(src,dst)
```

```python
from os import rename, makedirs

src='sunrise2.jpg'
dst='tmp/sunrise2.jpg'

makedirs('tmp', exist_ok=True) # Create the 'tmp' directory if it doesn't exist
rename(src,dst)
```

```python
from os import listdir

folder='./'
ret=listdir(folder)
print(ret)
```

```python
from glob import glob

target='*.txt'
ret=glob(target)
print(ret)
```

```python
target='**'
ret=glob(target,recursive=True)
ret.sort()
for r in ret:
  print(r)
```
