```python
data=['a','bc',1234,'def','ghij']
try:
  for d in data:
    assert type (d) is str, f'문자열이 아닌 데이터 <{d}> 있음!'
    print(d)
except AssertionError as e:
  print(e)
```

```python
def makeException():
  raise RuntimeError('뭔가 잘못되었음')
```

```python
try:
  makeException()
except Exception as e:
  print(e)
```

```python
try:
  makeException()
exception Exception as e:
  print(e)
  raise
```

```python
try:
  makeException()
exception Exception as e:
  raise ValueError('값이 잘못된 것 같음')
```
