```python
from os import listdir
from os.path import isdir

fileList=[]
dirList=[]
files=listdir('./')
for file in files:
    if isdir(file):
      dirList.append(f'<DIR> {file}')
    else:
      fileList.append(file)

fileList.sort()
dirList.sort()

for d in dirList:
  print(d)

for f in fileList:
  print(f)
```
