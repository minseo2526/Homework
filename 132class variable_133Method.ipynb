```python
class Wizard:
  generated=0

  def __init__(self, name):
    Wizard.generated += 1
    self.name = name
    self.level = 1
    self.hp = 100
    self.initAbility()


  def initAbility(self):
    self.strength=1
    self.offense_power=1
    self.intelligence=10
    self.mentality=10
    self.agility=1
```

```python
player1=Wizard('FirePlayer')
print(f'마법사생성 수:{player1.generated}\n캐릭터이름:{player1.name}')
```

```python
class Wizard:
  generated=0

  def setFireballLevel(self,power=1):
    self.fireballLevel=power
    print(f'화염구 강도를 {power}로 설정합니다.')

  @classmethod
  def addGenerated(cls,count):
    cls.generated+=count
    print(f'생성된 총 캐릭터 수:{cls.generated}')

  @staticmethod
  def static1():
    print('Wizard 클래스에서 정의된 정적 메서드입니다.')
```
