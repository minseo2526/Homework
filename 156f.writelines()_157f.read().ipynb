```python
data1='This is data1'
data2='This is data2'
with open('mydata.txt','w') as f:
  f.writelines(data1)
  f.writelines(data2)
```

```python
with open('mydata.txt','r') as f:
  data=f.read()
print(data)
```

```python
bufsize=1024
with open('sunrise.jpg','rb') as f:
  with open('sunrise_copied.jpg','wb') as h:
    data=f.read(bufsize)
    while data:
      h.write(data)
      data=f.read(bufsize)
```
